# Data Visualization

```python
df.head()
```

```python
# Price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df,x='Price_euros', bins=30, kde=True)
```

```python
# Price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Price_euros'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price (euros)')
plt.ylabel('Frequency')
plt.show()
```

```python
# RAM distribution
plt.figure(figsize=(10, 6))
sns.histplot(df,x='Ram', bins=10, kde=True)
plt.title('RAM Distribution')
plt.xlabel('RAM (GB)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Weight distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Weight'], bins=30, kde=True)
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

```python
lift_type_counts = df['TypeName'].value_counts()
# Plotting the pie chart
plt.figure(figsize=(12, 10))
plt.pie(lift_type_counts, labels=lift_type_counts.index, autopct='%1.1f%%', startangle=90, colors=['red','blue','green'])
plt.title('Distribution of TypeName')
plt.show()
```

```python
# price comparision by various cpu company types
plt.figure(figsize=(12, 6))
sns.lineplot(x='CPU_company', y='Price_euros', data=df,marker='o')
plt.title('Price Comparison by CPU company')
plt.xlabel('CPU company')
plt.ylabel('Price (euros)')
plt.grid(True)
plt.show()
```

```python
# price comparision by various laptop types
plt.figure(figsize=(12, 6))
sns.boxplot(x='TypeName', y='Price_euros', data=df)
plt.title('Price Comparison by Laptop Type')
plt.xlabel('Laptop Type')
plt.ylabel('Price (euros)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Price vs Ram
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Ram', y='Price_euros', hue='TypeName')# Colors the points based on the values in the 'TypeName' column. 
plt.title('Price vs. RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (euros)')
plt.legend(title='TypeName')
plt.show()
```

```python
#Screen Size vs. Price
plt.figure(figsize=(12, 8))
plt.scatter(df['Inches'], df['Price_euros'], alpha=0.7)  #sets the transparency of the data points.
plt.xlabel('Screen Size (Inches)')
plt.ylabel('Price (euros)')
plt.title('Price vs. Screen Size')
plt.show()
```

```python
g = sns.FacetGrid(df, col="TypeName", col_wrap=2, height=5)
#col_wrap=2 means that there will be up to 2 plots per row, and height=5 specifies the height of each subplot.
g.map_dataframe(sns.scatterplot, x='Weight', y='Price_euros')
g.set_axis_labels('Weight (kg)', 'Price (€)')
g.add_legend()
plt.show()
```

```python
# Creating the pair plot
sns.pairplot(df.select_dtypes(include=['number']))
plt.show()
```

```python
sns.pairplot(df[['Inches', 'Ram', 'Weight', 'Price_euros']])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x="Company", y="RetinaDisplay", data=df)

plt.xlabel("Company")
plt.ylabel("RetinaDisplay")
plt.title(" Distribution of Company")
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_matrix = df[['Inches', 'Ram', 'Weight', 'Price_euros']].corr()

# Generate heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
